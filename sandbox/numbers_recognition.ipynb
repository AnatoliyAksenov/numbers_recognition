```python
def dataset(file):
    label, gender, media = [], [], []
    with open(file, "r") as f:
        for line in f:
            l,g,m = line.split(',')
            label.append(l.strip())
            gender.append(g.strip())
            media.append(base64.b64decode(m))
    return label, gender, media
```
